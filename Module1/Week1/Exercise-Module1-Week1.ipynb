# Ex1

```python
def exercise1(tp, fp, fn):
    if(type(tp)!=int):
        print("tp is not an integer")
        return
    if(type(fp)!=int):
        print("fp is not an integer")
        return
    if(type(fn)!=int):
        print("fn is not an integer")
        return
    if(tp==0 or fp==0 or fn==0):
        print("tp and fp and fn must be greater than zero")
        return
    precision = tp / (tp + fp)
    recall = tp / (tp + fn)
    f1_score = 2 * precision * recall / (precision + recall)
    print(f"Precision is {precision}")
    print(f"Recall is {recall}")
    print(f"F1-score is {f1_score}")
```

```python
exercise1(2, 3, 4)
```

```python
exercise1('a', 3, 4)
```

```python
exercise1(2, 'b', 4)
```

```python
exercise1(2, 3, 'c')
```

```python
exercise1(2, 3, 0)
```

```python
exercise1(2.1, 3, 4)
```

```python
import random

def exercise3():
    n = input("Input number of samples ( integer number ) which are generated: ")
    if not n.isnumeric():
        print("number of samples must be an integer number")
        return
    n = int(n)
    loss_function = input("Input loss name: ")
    if loss_function not in ["MAE", "MSE", "RMSE"]:
        print(f"{loss_function} is not supported")
        return
    predicts = []
    targets = []
    losses = []
    for i in range(n):
        predicts.append(random.random() * 10)
        targets.append(random.random() * 10)
        if(loss_function == "MAE"):
            losses.append(math.abs(predicts[i] - targets[i]))
        else:
            losses.append((predicts[i] - targets[i])**2)
        print(f"loss name : {loss_function}, sample : {i+1}, pred : {predicts[-1]}, target : {targets[-1]}, loss : {losses[-1]}")
    final_loss = sum(losses) / n
    if(loss_function == "RMSE"):
        final_loss = math.sqrt(final_loss)
    print(f"final {loss_function}: {final_loss}")
```

```python
def fact(n):
    if n == 0:
        return 1
    return n * fact(n-1)
```

```python
def approx_sin(x, n):
    res = sum([(-1)**i * x**(2*i+1) / fact(2*i+1) for i in range(n)])
    return res
```

```python
def approx_cos(x, n):
    res = sum([(-1)**i * x**(2*i) / fact(2*i) for i in range(n)])
    return res
```

```python
def approx_sinh(x, n):
    res = sum([x**(2*i+1) / fact(2*i+1) for i in range(n)])
    return res
```

```python
def approx_cosh(x, n):
    res = sum([x**(2*i) / fact(2*i) for i in range(n)])
    return res
```

```python
approx_sin(3.14, 10)
```

```python
approx_cos(3.14, 10)
```

```python
approx_sinh(3.14, 10)
```

```python
approx_cosh(3.14, 10)
```

```python
def md_nre_single_sample(y, y_hat, n=2, p=1):
    return (y**(1/n) - y_hat**(1/n))**p
```

```python
md_nre_single_sample(100, 99.5)
```
